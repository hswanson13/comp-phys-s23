```python
import numpy as np

def func(x,u):
    #used sub u = y/2 -1
    y=2*(u+1)
    return (x+2*y)*(x+y)*2 #2 for dy=2*du

N=100000
xi = np.random.uniform(0,1,N)
ui = np.random.uniform(0,1,N)

monte = 1/N * np.sum(func(xi,ui))
#analytically it is exactly 47

err = np.std(func(xi,ui))/np.sqrt(N)
print('Area = 47 analytically')
print(f'Area = {monte} +/- {err}')
print(f'so it is within the error')
```
