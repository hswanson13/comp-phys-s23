```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
R = np.random.uniform(0,1,10**6) #uniform distrobution values of x

#https://www.cl.cam.ac.uk/teaching/0708/Probabilty/prob11.pdf
#found cdf of given pdf
#set it equal to R 
#inverted the function to find x

y = -0.25 + 0.75 * np.sqrt(1/9 + 8*R/3) #new distrobution of values that follows the given pdf

counts, bins, patches = plt.hist(y,bins=10)

bin_width = bins[1]-bins[0] #bin width
s = np.linspace(0,1,100)
integral = np.sum(counts*bin_width) #integral under curve

plt.plot(s, integral*(1+4*s)/3) #multiplied by the integral from the new found dist of values x

plt.show()
```
