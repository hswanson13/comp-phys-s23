```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#sum uniform dist til it is greater than 1
#the number of times added is n
#do this man times

#PART A
def get_X(trials):
    #trials = 10**6
    ns = np.zeros(trials)

    for i in range(trials):
        u = np.random.uniform(0,1,15)
        n = (np.cumsum(u)<=1).argmin() + 1
        if n==len(u):
            print('wow')
            print(i)
            continue #case where 15 random integers doesn't add to 1
        ns[i] = n
    return ns
```

```python
#PART B
ns = get_X(10**6)

b=np.arange(0,np.max(ns)+1,1)

plt.figure(figsize=(10,10))

x = np.linspace(3,10,100)

plt.hist(ns,bins=b)

plt.xticks(b)
plt.xlabel('n')
plt.ylabel('Counts')
plt.show()
```

```python

EX = np.mean(ns)
err = np.std(ns)

print(f'Mean: {EX}')
print(f'Error: {err}')
```

```python
X1 = get_X(10**5)
X2 = 1-X1

X = (X1+X2)/2

new_EX = np.mean(X)
new_err = np.std(X)

print(f'New Mean: {new_EX}')
print(f'New err: {new_err}')
```
