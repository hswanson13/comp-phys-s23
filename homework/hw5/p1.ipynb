```python
import numpy as np
from scipy import stats
```

```python
def pi(N):
    rand_x = np.random.uniform(0,1,N)
    rand_y = np.random.uniform(0,1,N)
    rand_r = np.sqrt(rand_x**2+rand_y**2)

    x_circ = rand_x[rand_r<1]
    y_circ = rand_y[rand_r<1]

    n = len(x_circ)
    return 4*n/N

Ns = [10**2,10**3,10**4,10**5,10**6]
chisqr = 0
for N in Ns:
    pie = pi(N)
    unc_f = np.sqrt((pie/4)*(1-(pie/4))/N)
    print(f'Total Number of points: {N}')
    print(f'pi_found: {pie}')
    print(f'pi_monte/pi_actual: {pie/np.pi}')
    pull = (pie-np.pi)/(4*unc_f)
    print(f'The pull: {pull}')
    chisqr+=pull**2
    print('')

    
    
print(chisqr/5)
prob = 1-stats.chi2.cdf(chisqr,5)
print(prob)
```
