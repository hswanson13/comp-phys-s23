```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,6*np.pi,100)

a = np.exp(-t/8)*np.sin(t)
b = np.exp(-t/8)*np.cos(t)

plt.plot(t,a, color='r')
plt.plot(t,b, color='b')
plt.plot(t,np.sqrt(a**2+b**2), color='g')

plt.xlabel('t')
plt.legend(['a','b','sqrt(a^2+b^2)'])

plt.show()
```
