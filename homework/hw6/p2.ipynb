```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#integrand = e^(x+y)*(x+y)^N * p(y)
#General MC integration = 1/N * sum[ f(y)/p(y) ] where y are points sampled from p
#Inside the sum is just: e^(x+y)*(x+y)^N, so that is all we need

#sample our gaussian, p(y)
samples=10**6
mu = 3
sigma = 0.5
p = np.random.normal(mu,sigma,samples)
N=5
def func(x,y,N):
    return np.exp(-(x+y))*(x+y)**N

points = 100
xs = np.linspace(-3,15,points)

f_x = np.zeros(points)
for i in range(points):
    f_x[i] = np.mean(func(xs[i],p,N))
```

```python
plt.figure(figsize=(8,6))
plt.plot(xs,f_x)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```
