```python
import numpy as np
import matplotlib.pyplot as plt
from decimal import *
```

```python
x = np.linspace(-1150,120,3500,dtype=float)

def plot_quad(x,a,b,c):
    quad_eq = ((-b+np.sqrt(b**2-4*a*(c)))/(2*a) , (-b-np.sqrt(b**2-4*a*(c)))/(2*a))
    print(f"Quadratic Eq: {quad_eq}")

    plt.plot(x,x**2+1000*x-1)
    plt.title('quadratic, x**2+1000*x-1')
    plt.xlabel('x')
    plt.ylabel('y=x**2+1000*x-1')
    plt.grid()
    
plot_quad(x,1,-1000,-1)
```

```python
#Newtons Method

def f_prime(x):
    return 2*x+1000

def f(x):
    return x**2+1000*x-1

def next_x(x_start,c,depth):
    c+=1
    x_curr = x_start - f(x_start)/f_prime(x_start)
    if c==depth:
        return x_curr
    return next_x(x_curr,c,depth)

#first time doing recursion!
print(f'Positive Root using Python float: {next_x(0.5,0,2000)}')

print(f'Nevative Root using Python float: {next_x(-1000.5,0,2000)}')
```

```python
#using decimal

plot_quad(x,Decimal(1),Decimal(-1000),Decimal(-1))
print('')
print(f"Positive Root using Decimal: {next_x(Decimal(0.5),0,2000)}")
print(f"Negative Root using Decimal: {next_x(Decimal(-1000.5),0,2000)}")
```

Interestingly enough, using the Decimal package to calculate the quadratic equation
gave an answer that reflected that of my root finding algorithm. Decimal handles floating point arithematic better than a normal float, so that is probably why it worked better than calculating the roots using the quadratic formula outright. Floats performed fine in the algorithim because every iteration I was gaining more precision in my iterative process.
