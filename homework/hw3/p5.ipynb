```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x) - (0.4/x) * np.cos(x)

def root(xL,xR,depth):
    plot_params = np.empty([depth,2]) #depth is the number of arrays, 2 is number of elements in the array
    for i in range(depth):
        
        #m = (f(xR)-f(xL))/(xR-xL)
        plot_params[i] = np.array([xL,xR])
        
        x1 = xR - f(xR)*float(xR-xL)/float(f(xR)-f(xL))
        xL = xR  #each time x1 steps closer to the zero point until it finally reaches the end
        xR = x1
    return plot_params


rs = root(0.3, 1, 4)

rs2 = root(2,4,4)

rs3 = root(5,7.5,4)
```

```python
def plot_progress(x,rs,rnum):
    print('Colors are a bit confusing, use reference numbers to see how the iterstions go')
    c=0
    for p in rs[::1]:
        #print(f'actual {p[1]}')
        c+=1
        m = (f(p[1])-f(p[0]))/(p[1]-p[0])

        xint = p[1] - f(p[1])*(p[1]-p[0])/(f(p[1])-f(p[0]))
        print(f'xint {xint}, c={c}')
        y_xint = m*(xint-p[1])+f(p[1])

        y = m*(x-p[1])+f(p[1])

        plt.plot(x,y, label = f'Iteration Number {rnum} {c}')
        plt.plot(xint, y_xint, marker='.')
        plt.legend()
        
        
xlimit = 8
x = np.linspace(0.15,xlimit,300)

plt.figure(figsize=(10,10), dpi=80)

plt.plot(x,f(x))


plot_progress(x,rs,1)
plot_progress(x,rs2,2)
plot_progress(x,rs3,3)


    
plt.xlim([0, xlimit])
plt.ylim([-3,2])
plt.title('Secant Root Finding Technique Progression')
plt.xlabel('x')
plt.ylabel('sin(x) - (0.4/x)cos(x)')
plt.grid()
```
