```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
x = np.linspace(0,1,num=1000)
y = np.linspace(0,1,num=1000)
#print(x)
def psi2(x,y):
    A = 2 #normalization constant 2/L
    L = 1
    nx = 2
    ny = 4
    psi = A*np.sin(nx*np.pi*x/L)*np.sin(ny*np.pi*y/L)
    return psi**2 #just square is fine here, no imaginary parts in sight!

space_map = np.array([psi2(x,i) for i in y])

ax = sns.heatmap(space_map)

ax.set_xticklabels(ax.get_xticks()/1000)
ax.set_yticklabels(ax.get_yticks()/1000)

ax.invert_yaxis()
plt.show()
```
