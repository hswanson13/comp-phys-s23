```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
x = np.linspace(0,1,1000)
y = np.linspace(0,1,1000)

def psi2(x,y):
    A = 2 #normalization constant 2/L
    L = 1
    nx = 2
    ny = 3
    psi = A*np.sin(nx*np.pi*x/L)*np.sin(nx*np.pi*y/L)
    return psi**2 #just square is fine here, no imaginary parts in sight!

space_map = np.array([psi2(x,i) for i in y])

ax = sns.heatmap(space_map)
ax.invert_yaxis()
```
