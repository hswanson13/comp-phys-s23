```python
import matplotlib.pyplot as plt
import numpy as np
from plotting import statBox, plotErr
```

```python

file_dir = '/home/hayden/comp401/frank_repo/comp-phys-s23/homework/hw2/dataSet.npy'
x = np.load(file_dir)

fig, ax = plt.subplots(ncols=1, figsize=(8, 4))


bins = 250
plot = ax.hist(x, bins)

binEdges = np.histogram_bin_edges(x, bins=bins, range=None, weights=None)
#print(binEdges)

#print(len(binEdges),len(plot[2]))

statBox(ax,x,binEdges)
plotErr(ax,plot[0],binEdges) #second argument is the values of the histogram bins, counts

plt.yscale('log')
plt.title('Histogram of x')
plt.ylabel('Counts')
plt.xlabel('x')
plt.show()
```
